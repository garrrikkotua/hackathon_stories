```python
import numpy as np
import pandas as pd
```

```python
trans = pd.read_csv('transactions.csv', header=0, sep=',')
```

```python
trans.head()
```

```python
mccs = trans.merchant_mcc.value_counts()
```

```python
mccs.sum()
```

```python
mccs[:10].sum()
```

```python
new_mccs = np.array(mccs[:10].index)
```

```python
new_mccs
```

```python
trans = trans[np.in1d(np.array(trans.merchant_mcc), np.array(mccs[:10].index))]
```

```python
trans = trans.drop(['transaction_month', 'transaction_day', 'merchant_id'], axis=1)
```

```python
transgb = trans.groupby(['customer_id', 'merchant_mcc']).agg(['count', 'mean'])
```

```python
index = sorted(np.array(trans.customer_id.value_counts().index))
```

```python
new_mccs.sort()
```

```python
columns = np.array([str(x) + '_cnt' for x in new_mccs] + [str(x) + '_mean' for x in new_mccs])
columns
```

```python
new_trans = pd.DataFrame(index=index, columns=columns)
```

```python
new_trans = new_trans.fillna(0)
```

```python
transgb
```

```python
for num, (index, row) in enumerate(transgb.iterrows()):
    if num % 1000 == 0:
        print(num)
    cust_id, merch_mcc = index
    cnt, mean = row
    new_trans.loc[cust_id, str(merch_mcc)+'_cnt'] = cnt
    new_trans.loc[cust_id, str(merch_mcc)+'_mean'] = mean
    
```

```python
new_trans
```

```python
new_trans.to_csv('transformed_transactions.csv', sep=',', index=True, index_label='customer_id')
```

```python
custs = pd.read_csv('customer_train.csv', sep=',', header=0, parse_dates=['first_session_dttm'])
custs.head()
```

```python
custs.drop('job_title', axis=1, inplace=True)
```

```python
custs['full_years'] = custs.first_session_dttm.apply(lambda date: 2019 - date.year)
```

```python
custs.loc[0, 'first_session_dttm'].year
```

```python
custs.drop('first_session_dttm', axis=1, inplace=True)
```

```python
custs
```

```python
custs[['age', 'full_years']] = custs[['age', 'full_years']].fillna(custs[['age', 'full_years']].mean())
```

```python
custs.info()
```

```python
custs.fillna('NaN', inplace=True)
```

```python
custs
```

```python
custs['gender'] = custs.gender_cd.apply(lambda x: 1 if x == 'M' else 0)
```

```python
custs.drop('gender_cd', axis=1, inplace=True)
```

```python
custs
```

```python
cat_features = ['product_0', 'product_1', 'product_2', 'product_3', 'product_4', 'product_5', 'product_6',
                'marital_status_cd', 'job_position_cd']
```

```python
custs = pd.concat([custs] + [pd.get_dummies(custs[col], col) for col in cat_features], axis=1).drop(cat_features, axis=1)
```

```python
custs['children_info'] = custs.children_cnt.apply(lambda x: 0 if x == 'NaN' else 1)
custs['children_num'] = custs.children_cnt.apply(lambda x: 0 if x == 'NaN' else x)
```

```python
custs.drop('children_cnt', axis=1, inplace=True)
```

```python
custs
```

```python
new_trans = pd.read_csv('transformed_transactions.csv', sep=',', header=0)
new_trans
```
